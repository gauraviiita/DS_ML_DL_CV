# Canny edge detection

Apply Gaussian filter to smooth the image in order to remove the noise

find the intensity gradient in the image

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread("../DATA/sammy_face.jpg")
```

```python
plt.imshow(img)
```

```python
edges = cv2.Canny(image=img, threshold1=127, threshold2=127)
```

```python
plt.imshow(edges)
```

```python
edges = cv2.Canny(image=img, threshold1=0, threshold2=255)
plt.imshow(edges)
```

```python
med_val = np.median(img)
```

```python
med_val
```

```python
# low threshold to either 0 or 70% of the median value whichever is greater
lower = int(max(0, 0.7*med_val))

#upper threshold to either 130% of the median or the max 255, whichever is smaller
upper = int(min(255, 1.3*med_val))
```

```python
edges = cv2.Canny(image=img, threshold1=lower, threshold2=upper+100)
```

```python
plt.imshow(edges)
```

```python
blurred_img = cv2.blur(img, ksize=(5,5))
```

```python
edges = cv2.Canny(image = blurred_img, threshold1=lower, threshold2=upper+50)
plt.imshow(edges)
```
